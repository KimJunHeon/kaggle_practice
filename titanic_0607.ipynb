```python
train_y = pd.DataFrame(data['Survived'])
```

```python
train_x = data
train_x.drop(columns=['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin'], inplace = True)
```

```python
sex_list = data['Sex'].tolist()
is_male = list()

for sex in sex_list:
    if(sex == 'male'):
        is_male.append(1)
    else:
        is_male.append(0)
        
train_x.drop(columns=['Sex'], inplace = True)
train_x['IS_MALE'] = is_male
```

```python
train_x.head(10)
```

```python
# 남, 녀의 평균 나이 구하기
average_for_male = round(train_x[train_x.Age.notnull()][train_x['IS_MALE'] == 1]['Age'].mean())
average_for_female = round(train_x[train_x.Age.notnull()][train_x['IS_MALE'] == 0]['Age'].mean())
```

```python
# 나이 결측치는 성별의 평균으로 채워주기
mask_for_male = train_x.Age.isnull() & train_x.IS_MALE == 1
train_x.loc[mask_for_male, 'Age'] = average_for_male
mask_for_female = train_x.Age.isnull()
train_x.loc[mask_for_female, 'Age'] = average_for_female
```

```python
train_x.head(10)
```

```python
train_x['Embarked'].unique()
```

```python
# embarked에 대해서 encoding

embarked_list = train_x['Embarked'].tolist()
is_embarked_S = list()
is_embarked_C = list()
is_embarked_Q = list()

for embark in embarked_list:
    if(embark == 'S'):
        is_embarked_S.append(1)
        is_embarked_C.append(0)
        is_embarked_Q.append(0)
    elif(embark == 'C'):
        is_embarked_S.append(0)
        is_embarked_C.append(1)
        is_embarked_Q.append(0)
    elif(embark == 'Q'):
        is_embarked_S.append(0)
        is_embarked_C.append(0)
        is_embarked_Q.append(1)
    else:
        is_embarked_S.append(0)
        is_embarked_C.append(0)
        is_embarked_Q.append(0) 
        
train_x.drop(columns=['Embarked'], inplace = True)
train_x['IS_EMBARKED_S'] = is_embarked_S
train_x['IS_EMBARKED_C'] = is_embarked_C
train_x['IS_EMBARKED_Q'] = is_embarked_Q        


# Pclass에 대해서 Encoding
pclass_list = train_x['Pclass'].tolist()
is_pclass_1 = list()
is_pclass_2 = list()

for pclass in pclass_list:
    if(pclass == 1):
        is_pclass_1.append(1)
        is_pclass_2.append(0)
    elif(pclass == 2):
        is_pclass_1.append(0)
        is_pclass_2.append(1)
    else:
        is_pclass_1.append(0)
        is_pclass_2.append(0)
        
train_x.drop(columns=['Pclass'], inplace = True)
train_x['IS_PCLASS_1'] = is_pclass_1
train_x['IS_PCLASS_2'] = is_pclass_2
```

# Modeling

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 501, criterion='gini', n_jobs = -1, random_state = 1)
model.fit(train_x, train_y)
prediction = model.predict(test_x)
```
